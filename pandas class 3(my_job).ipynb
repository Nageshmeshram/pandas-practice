```python
data4 = {'emp_id' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }

data5  = {'emp_id' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
```

```python
df4
```

```python
df5
```

```python
pd.merge(df4,df5)
```

```python
pd.merge(df4,df5,how = 'left') # it is 1st copy data  from 1st table then fill data from 2nd table
```

```python
pd.merge(df4,df5,how = 'right') # it is first second table then fill data from 1st table
```

```python
pd.merge(df4,df5,how = 'inner') # it is by defoult inner merge
```

```python
pd.merge(df4,df5,how = 'cross') # it will multipy one table with another 
```

```python
pd.merge(df4,df5,on = 'emp_id')
```

```python
def profit_flag(a):
    if a>0:
        return "Positive"
    else :
        return "Negative"
    
```

```python
profit_flag(-10)
```

```python
profit_flag(10)
```

```python
sale_data['flag profit']=sale_data['profit'].apply(profit_flag)
```

```python
sale_data.head(10)
```

```python
sale_data.drop('flage profit',axis=1,inplace=True)
```

```python
sale_data
```

```python
sale_data['len of cust name']=sale_data['customer_name'].apply(len)
```

```python
sale_data
```

```python
def quantity_flag(a):
    if a<5:
        return "LOW"
    elif a>5 and a<8:
        return "MEDIUM"
    else:
        return "HIGH"
```

```python
sale_data['flag quantity']=sale_data['quantity'].apply(quantity_flag)
```

```python
sale_data['square_quantity']=sale_data['quantity'].apply(lambda a:a*a)
```
